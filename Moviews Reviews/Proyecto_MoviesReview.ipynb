```python
dtMovies['critic_sentiment'] = dtMovies['critic_sentiment'].replace({'positive': 1, 'negative': 0})
```

```python
dtCriticReviews.head(5)
```

```python
# prompt: join  dtCriticReviews with dtMovies where we will need only the column title

dtCriticReviewsMovie = dtCriticReviews.merge(dtMovies[['movieTitle','movieId']], on='movieId', how='left')
dtCriticReviewsMovie.head()
```

```python
dtCriticReviews.columns
```

```python
# Sentiment distribution
dtCriticReviews['isRotten'].value_counts().plot(kind='bar')
plt.xlabel('Rotten')
plt.ylabel('Count')
plt.title('Sentiment Distribution')
plt.show()
```

```python
# prompt: plot to find the top movies with positive sentiment

import matplotlib.pyplot as plt
# Group by movie title and count positive reviews
positive_reviews = dtCriticReviewsMovie[dtCriticReviewsMovie['isFresh'] == True].groupby('movieTitle')['isFresh'].count().sort_values(ascending=False)

# Plot top 10 movies with positive sentiment
positive_reviews[:10].plot(kind='bar')
plt.xlabel('Movie Title')
plt.ylabel('Number of Positive Reviews')
plt.title('Top 10 Movies with Positive Sentiment')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

# Group by movie title and count positive reviews
positive_reviews = dtCriticReviewsMovie[dtCriticReviewsMovie['isFresh'] == False].groupby('movieTitle')['isFresh'].count().sort_values(ascending=False)

# Plot top 10 movies with positive sentiment
positive_reviews[:10].plot(kind='bar')
plt.xlabel('Movie Title')
plt.ylabel('Number of Negative Reviews')
plt.title('Top 10 Movies with Positive Sentiment')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Sentiment distribution
dtCriticReviews['isFresh'].value_counts().plot(kind='bar')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Sentiment Distribution')
plt.show()
```

```python
# Top movies by number of reviews
dtCriticReviewsMovie['movieTitle'].value_counts().head(10).plot(kind='barh')
plt.xlabel('Number of Reviews')
plt.ylabel('Critic')
plt.title('Top 10 Movies by Number of Reviews')
plt.show()
```

```python

# Top critics by number of reviews
dtCriticReviews['quote'].value_counts().head(10).plot(kind='barh')
plt.xlabel('Number of Reviews')
plt.ylabel('Critic')
plt.title('Top 10 Critics by Number of Reviews')
plt.show()


```

```python
# Publication distribution
dtCriticReviews['publicationName'].value_counts().head(10).plot(kind='barh')
plt.xlabel('Number of Reviews')
plt.ylabel('Publication')
plt.title('Top 10 Publications by Number of Reviews')
plt.show()
```

```python
dtUsersReviewMovie = dtUserReviews.merge(dtMovies[['movieTitle','movieId']], on='movieId', how='left')
dtUsersReviewMovie.head()
```

```python

# Top 10 movies by average user rating
dtUsersReviewMovie.groupby('movieTitle')['rating'].mean().sort_values(ascending=False).head(10).plot(kind='barh')
plt.xlabel('Average Rating')
plt.ylabel('Movie Title')
plt.title('Top 10 Movies by Average User Rating')
plt.show()

# Number of reviews per movie
dtUsersReviewMovie['movieTitle'].value_counts().head(10).plot(kind='barh')
plt.xlabel('Number of Reviews')
plt.ylabel('Movie Title')
plt.title('Top 10 Movies by Number of User Reviews')
plt.show()

```

```python
# Distribution of review lengths (word count)
dtUserReviews['review_length'] = dtUserReviews['quote'].apply(lambda x: len(x.split()))
plt.hist(dtUserReviews['review_length'], bins=50)
plt.xlabel('Review Length (Words)')
plt.ylabel('Frequency')
plt.title('Distribution of User Review Lengths')
plt.show()
```
